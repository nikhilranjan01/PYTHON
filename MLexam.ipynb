```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Sample data
data = np.array([[2, 2, 0],
                 [1, 2, 0],
                 [3, 4, 0],
                 [1, 1, 0],
                 [3, 3, 0],
                 [7, 2, 1],
                 [6, 2, 1],
                 [7, 1, 1],
                 [8, 0, 1],
                 [7, 3, 1]])


X = data[:, :2]
Y = data[:, 2]


knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, Y)


plt.scatter(X[Y == 0][:, 0], X[Y == 0][:, 1], c='b', label='Class 0')
plt.scatter(X[Y == 1][:, 0], X[Y == 1][:, 1], c='r', label='Class 1')


test_data = np.array([[3.5, 4.5]])
plt.scatter(test_data[:, 0], test_data[:, 1], c='g', label='Test Data (3.5, 4.5)', marker='x', s=100)


predicted_class = knn.predict(test_data)
if predicted_class == 0:
    plt.title(f'Test Data: Class 0')
else:
    plt.title(f'Test Data: Class 1')

plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
